# Some Simple Shapes

```python
def square(size):
    t = Turtle()
    for i in range(4):
        t.forward(size)
        t.left(90)
        
def triangle(size):
    t = Turtle()
    for i in range(3):
        t.forward(size);
        t.right(120);
        
square(50)
triangle(50)
```

```python
def polystop(t,side, angle):
    turn = 0
    while True:
        t.forward(side)
        t.right(angle)
        turn += angle
        if turn % 360 == 0:
            break
t = Turtle()
polystop(t,50,60)
```

# Some Odder Figures

```python
def thing(t):
    steps = [100,100,50,50,100,25,25,50];
    for s in steps[:-1]:
        t.forward(s);
        t.right(90);
    t.forward(steps[-1]);


def thing1():
    t = Turtle()
    for i in range(4):
        thing(t)
        
thing1()
```

```python
def arcr(t, r, deg):
    for i in range(deg):
        t.forward(r);
        t.right(1);

def arcl(t, r, deg):
    for i in range(deg):
        t.forward(r);
        t.left(1);
        
def circles():
    t = Turtle()
    for i in range(9):
        arcr(t,1,360);
        t.right(40);

circles()
```

```python
def petal(t,size):
    arcr(t,size,60)
    t.right(120)
    arcr(t,size,60)
    t.right(120)

def flower(n=6):
    size = 2
    t = Turtle()
    for i in range(n):
        petal(t,size)
        t.right(360/n)
flower(12)
```

# Some Fractals

```python
def nested_triangle(t,size):
    if size < 5:
        return
    for i in range(3):
        nested_triangle(t,size/2)
        t.forward(size)
        t.right(120)
t = Turtle()
t.left(30)
nested_triangle(t,400)
t.clip(border=5)
```

```python
def branch(t,length, level):
    if level == 0:
        return
    t.forward(length)
    t.left(45)
    branch(t,length/2, level-1)
    t.right(90)
    branch(t,length/2,level-1)
    # return to original state
    t.left(45)
    t.back(length)
t = Turtle()
t.left(90)
branch(t,200,6)
t.clip(border=5)
```

```python
def weird_tree(t,length,angle,level):
    def lbranch(length, angle, level):
        t.forward(2*length)
        node(length, angle, level)
        t.back(2*length)

    def rbranch(length, angle, level):
        t.forward(length)
        node(length, angle, level)
        t.back(length)

    def node(length, angle, level):
        if level == 0:
            return
        t.left(angle)
        lbranch(length, angle, level-1)
        t.right(2*angle)
        rbranch(length,angle,level-1)
        t.left(angle)
    lbranch(length,angle,level)
t = Turtle()
t.left(90)
weird_tree(t,30,20,7)
t.clip(border=5)
```

```python
def koch_snowflake(t,size):
    if size < 5:
        t.forward(size)
        return
    koch_snowflake(t,size/3)
    t.left(45)
    koch_snowflake(t,size/3)
    t.right(90)
    koch_snowflake(t,size/3)
    t.left(45)
    koch_snowflake(t,size/3)
t = Turtle()
koch_snowflake(t,200)
t.clip(border=5)
```
